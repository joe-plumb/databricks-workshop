# Mount Azure Blob Storage on Azure Databricks

## Create a Storage account in the Azure Portal (Class)
Capture the St

```
Storage_account = "{storage account name}"
Storage_container = "{container name}"
Storage_key = "{storage access key}"
```

#### Mount blob storage to a DBFS mount

```
source_string = "wasbs://"+Storage_container+"@"+Storage_account+".blob.core.windows.net/"
config_string = "fs.azure.account.key."+Storage_account+".blob.core.windows.net"
dbutils.fs.mount(
  source = source_string,
  mount_point = "/mnt/blobmount",
  extra_configs = {config_string:Storage_key})
```

#### List Mounts

```
dbutils.fs.mounts()
```

```
%fs
ls /mnt/blobmount
```

#### Using the Fuse Mount
Databricks also provides a "fuse mount" /dbfs that allows posix access to the blob storage. 
Use with caution, small files only <1GB

```
%sh
ls /dbfs/mnt/blobmount/
```

```
%sh
echo "hello world" > /dbfs/mnt/blobmount/helloworld.txt
ls -lt /dbfs/mnt/blobmount/helloworld.txt
cat /dbfs/mnt/blobmount/helloworld.txt
# rm /dbfs/mnt/blobmount/helloworld.txt
```

### DB Mounts Unmounting
By default DBFS mounts are persisted across cluster reboots, however they can be manually umounted

```
dbutils.fs.unmount("/mnt/blobmount")
```
